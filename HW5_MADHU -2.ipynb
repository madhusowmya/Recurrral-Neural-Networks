```python
data = pd.read_csv('/content/drive/My Drive/Colab Notebooks/HW5/airline-passengers.csv')
data
```

```python
data['Month'] = pd.to_datetime(data['Month'])
```

```python
data.info()
```

```python
data
```

```python
plt.xlabel('Month')
plt.ylabel('No of Passengers')
plt.plot(data['Passengers'])
plt.title("Time series data")
```

```python
data['Month-str'] = data['Month'].dt.strftime('%b')
data['year'] = data['Month'].dt.year
data.head(10)
```

```python
data['year'] = data['year'].astype(str)
data['month'] = data['Month-str'].astype(str)
data['Month-Year'] = data['Month-str'] + " - " + data['year']
```

```python
data[data['year']<'1961'].set_index('Month-Year')['Passengers'].plot(kind='bar',
                                                                                figsize=(50, 10),
                                                                                 color='cadetblue',
                                                                               )
plt.title("Passengers Per Month Yearwise", y=1.013, fontsize=22)
plt.ylabel("Number of Passengers", labelpad=16)
plt.xlabel("month - year", labelpad=16);
```

Part b)



From the above plot depicting Number of passengers vs Month and Year for the years (1949 to 1960), It can be observed that The trend of the number of passengers is increasing for the months between DECEMBER and MARCH, The number passengers is decreasing after March and again improves July, After July the number of passengers decreases again until November.
Long term temporal pattern for increase in the number of passengers is between NOVEMBER and MARCH
Short term temporal pattern for increase in the number of passengers is between JUNE and AUGUST
